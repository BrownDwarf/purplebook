```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def cp(T):
    A,B,C,D,E=ShomateCoeffs
    return A + B*(T/1000) + C*(T/1000)**2 + D*(T/1000)**3 + E*(T/1000)**-2
```

```python
R_starred = 8314.5 #(m/s)^2 /K
M = 44 #g/mol
#1 mb = 100 Pa
```

```python
CO2_coeffs = [568.122, 1254.249, -765.713, 189.645, -3.105]
ShomateCoeffs = CO2_coeffs
```

```python
R = R_starred/M

# Define a function which calculates the derivative
def f(x, Y):
    return R/cp(np.exp(Y))

a=np.log(92000*100) #start of the interval                                                  
b=np.log(100*100) # end of the interval (10 minutes)
N= 1000 #number of steps
h=(b-a)/N #size of an individual step                                          

##Euler's method                                                                                                                                   
Y=np.log(737) #initial condition
 
xpoints=np.arange(a,b,h)

Y_solution=[]

for x in xpoints:
    Y_solution.append(Y)
    Y= Y+h*f(x, Y)

print('The initial temperature at a pressure of 100 mb is', np.exp(Y_solution[0]), 'K.')
print('The final temperature at a pressure of 100 mb is', np.exp(Y_solution[-1]), 'K.')
```

```python
plt.ylabel("T (K)")
plt.xlabel("p (bars)")
plt.xscale('log')
plt.gca().invert_xaxis()
plt.plot(np.exp(xpoints)/10**5, np.exp(Y_solution))
plt.show()
```

```python
plt.xlabel("T (K)")
plt.ylabel("p (bars)")
plt.yscale('log')
plt.gca().invert_yaxis()
plt.plot(np.exp(Y_solution), np.exp(xpoints)/10**5)
plt.show()
```
